```python
import pandas as pd
import textwrap
from bs4 import BeautifulSoup
import requests
from tqdm import tqdm
import time
from urllib.parse import urljoin
```

```python
BaseURL = "https://pford.info/refolddb/"
search_all = "./list.cgi?param=*&lang=EN"

search_news = "./list.cgi?kind=208&param=non_aggregability&lang=EN"

search_all_url = urljoin(BaseURL,search_all)

search_news_url = urljoin(BaseURL, search_news)

r = requests.get(search_news_url)

soup = BeautifulSoup(r.text, "html.parser")
links = soup.find_all("a",string="Detail")
```

```python
def FindMonashUrl(url):
    
    # Get the page
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")

    # Find all the table rows
    rows = soup.find_all("tr")

    for row in rows:
        # checks if one of the column is "Link"
        if row.find("td", string="Link") != None:
            # get the url
            return row.find("a").get("href")
    
    return None
```

```python
def FindDOI(url):

    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")

    return soup.find("meta", attrs={"name": "citation_doi"}).get("content")
```

```python
def FindPubMedURL(url):

    # Get the page
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "html.parser")

    # Find all the table rows
    rows = soup.find_all("tr")

    for row in rows:
        # checks if one of the column is "Link"
        if row.find("td", string="PubMed ID") != None:
            # get the url
            return row.find("a").get("href")
    
    return None
```

```python

prot_url = urljoin(BaseURL, links[0].get("href"))


for link in links:
    prot_url = urljoin(BaseURL, link.get("href"))
    if FindMonashUrl(prot_url) is None:
        print(FindDOI(FindPubMedURL(prot_url)))
    else:
        print(FindMonashUrl(prot_url))
```
